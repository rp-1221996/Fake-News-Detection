## Clean up text

```python
def clean_text(text):
    text = "".join([word for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopwords]
    return text

dataset['text_nostop'] = dataset['text'].apply(lambda x: clean_text(x.lower()))

dataset.head()
```

## Stem the Text

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

dataset['text_stemmed'] = dataset['text_nostop'].apply(lambda x: stemming(x))

dataset.head()
```

## Create function to remove punctuation, tokenize, remove stopwords and stem

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join([ps.stem(word) for word in tokens if word not in stopwords])
    return text

dataset['cleaned_text'] = dataset['text'].apply(lambda x: clean_text(x))
dataset.head()
```

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score
```

```python
X = dataset['cleaned_text']
y = dataset['label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
y_train
```

## Apply TfidfVectorizer

```python
tfvect = TfidfVectorizer(stop_words='english',max_df=0.7)
tfid_x_train = tfvect.fit_transform(x_train)
tfid_x_test = tfvect.transform(x_test)
```

```python
classifier = PassiveAggressiveClassifier(max_iter=50)
classifier.fit(tfid_x_train,y_train)
```

```python
y_pred = classifier.predict(tfid_x_test)
score = accuracy_score(y_test,y_pred)
print(f'Accuracy: {round(score*100,2)}%')
```

```python
import pickle
pickle.dump(classifier,open('model.pkl', 'wb'))
```
